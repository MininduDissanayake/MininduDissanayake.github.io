##Activity #1

Derive the 1-dof freedom model for describing the small amplitude motion of the system.

Analyzing the moment about the pivot, the equation of motion is given by:

\begin{align}
I_A\ddot{\theta} + kx_s l_{\text{spring}} = 0
\end{align}

The total mass moment of inertia for the complete assembly, denoted as \(I_A\), is the sum of the individual contributions from the beam, the mass, and the spring:

\begin{align}
I_A = I_{\text{beam}} + I_{\text{exciter}} + I_{\text{spring}}
\end{align}

For the beam's moment of inertia (\(I_{\text{beam}}\)):

\begin{align}
I_{\text{beam}} = \frac{1}{12}m_{\text{beam}}l_{\text{beam}}^2 + m_{\text{beam}}\left(\frac{l_{\text{beam}}}{2}\right)^2 = \frac{m_{\text{beam}}l_{\text{beam}}^2}{3}
\end{align}

For the exciter's moment of inertia (\(I_{\text{exciter}}\)):

\begin{align}
I_{\text{exciter}} = m_{\text{exciter}}l_{\text{exciter}}^2
\end{align}

Utilizing the small angle approximation \(X_s = \theta l_{\text{spring}}\), the equation simplifies to:

\begin{align}
I_A\ddot{\theta} + kl_{\text{spring}}^2 = 0
\end{align}

This equation can be expressed as:

\begin{align}
\ddot{\theta} + \left(\frac{kl_{\text{spring}}^2}{I_A}\right)\theta = 0
\end{align}

The general solution to this second-order differential equation is:

\begin{align}
\theta = A\cos\left(\sqrt{\frac{kl_{\text{spring}}^2}{I_A}}t - \alpha\right)
\end{align}

Here, \(A\) represents the amplitude, \(t\) is time, and \(\alpha\) is the phase angle determined by initial conditions.

##Activity #2

Estimate the damped natural frequency, $\omega_d$, and the amplitude of reseonance, $x_{ss_{max}}$, for small amplitude vibrations.

# **Free Vibration Analysis**

Consider a system with a damper attached to a beam and spring. The equation of motion is given by:

\begin{align}
I_A\ddot{\theta} = -kx_sl_{\text{spring}} - cl_{\text{damper}}\dot{x}_d
\end{align}

Rearranging, we obtain:

\begin{align}
\ddot{\theta} + \frac{cl_{\text{damper}}^2}{I_A}\dot{\theta} + \frac{kl_{\text{spring}}^2}{I_A}\theta = 0
\end{align}

Introduce the decay coefficient (\(\gamma\)) to represent damping:

\begin{align}
\ddot{\theta} + 2{\gamma}\dot{\theta} + {\omega}^2{\theta} = 0
\end{align}

Where:

\begin{align}
{\gamma} = \frac{c\ l_{\text{damper}}^2}{2I_A}
\end{align}

Define the damped natural frequency (\({\omega}_d\)) as:

\begin{align}
{\omega}_d = \sqrt{{\omega}^2-{\gamma}^2}
\end{align}

The amplitude of resonance (\({\theta}_{ss_{\text{max}}}\)) is given by:

\begin{align}
{\theta}_{ss_{\text{max}}} = \frac{{\tau}_0 / I_A}  {2 \gamma \omega_n}
\end{align}

If the amplitude of the driving force (\({\theta}_0 = 0\)), then:

\begin{align}
{\theta}_{ss_{\text{max}}} = 0
\end{align}

# **Forced Vibration Analysis**

For forced vibration with an applied force \(I_{A}\ddot{\theta} + c\ l_{\text{damper}}^2\dot{\theta} = I_{\text{mass}}Q\sin{\Omega}t\), where:

\begin{align}
Q = {\Omega}^2m_{\text{offsetmass}}r_{\text{offsetmass}}
\end{align}

The equation becomes:

\begin{align}
\ddot{\theta} + 2{\gamma}\dot{\theta} + {\omega}^2{\theta} = q\sin{\Omega}t
\end{align}

Here, \(q = \frac{l_{\text{exciter}}Q}{I_A}\).

The solution is given by:

\begin{align}
{\theta} = M \cos{\Omega}t + N \sin{\Omega}t
\end{align}

Alternatively,

\begin{align}
\theta = \frac{q}{{\omega}^2}{\beta  \sin({\Omega}t-\phi})
\end{align}

Where:

\begin{align}
\beta = \frac{1}{\sqrt{(1-{\Omega}^2/{\omega}^2)^2 + (2\zeta{\Omega}/{\omega})^2}}
\end{align}

And the phase lag (\(\phi\)) of the offset mass is given by:

\begin{align}
\tan(\phi) = \frac{2\zeta{\Omega}/{\omega}}{1-{\Omega}^2/{\omega}^2}
\end{align}

Finally, the amplitude of resonance (\({\theta}_{ss_{\text{max}}}\)) remains the same:

\begin{align}
{\theta}_{ss_{\text{max}}} = \frac{{\tau}_0 / I_A}  {2 \gamma \omega_n}
\end{align}

Consider the quadratic equation:

\begin{equation}
r^2 + 2 \gamma r + \omega^2 = 0
\end{equation}

The solutions to this equation are given by:

\begin{equation}
r = -\gamma \ \pm \ \sqrt{\gamma^2 - \omega^2}
\end{equation}

Now, let's explore the three cases that arise:

1. For the underdamped case, where $\gamma < \omega$, there are two complex roots:

\begin{align}
r_1 & = -\gamma \ + \ i\omega_d\\
r_2 & = -\gamma \ - \ i\omega_d
\end{align}

Incorporating these roots into the general solution $\theta = Ce^{rt}$ and expressing it in phase form, we obtain:

\begin{equation}
\theta = A e^{-\gamma t} \cos (\sqrt{(\omega^2 - \gamma^2)}\ t-\alpha_d)
\end{equation}

Here, $A$ and $\alpha_d$ represent the amplitude and phase, respectively, determined by initial conditions.

2. In the critically damped case, where $\gamma = \omega$, there is a single real root.

3. In the overdamped case, where $\gamma > \omega$, there are two real roots.

Additionally, for the underdamped case, the following relationships hold:

\begin{equation}
\sqrt{\omega^2 - \gamma^2} \tau_d = 2\pi = \frac{2\pi}{\omega_d}
\end{equation}

\begin{equation}
\omega_d^2 = \omega^2 - \gamma^2
\end{equation}

\begin{equation}
\frac{\omega_d}{\omega} = \sqrt{1-\zeta^2}
\end{equation}

Here, $\tau_d$ represents the damping time constant, $\omega_d$ is the damped angular frequency, and $\zeta$ is the damping ratio.

##Activity #3

Plot the following:

* the unforced repsonse to an initiail diplacement of the spring end of the beam
* the forced response for three choices of the rotational speed of the motor
* theoretical frequency response of the system

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define system parameters
mass_system = 1.0
pivot_distance = 1.0
motor_arm_length = 1.0
moment_of_inertia = 1.0

# Undamped Natural Frequency
spring_stiffness = 100.0
spring_length = 1.0
undamped_natural_frequency = np.sqrt((spring_stiffness * spring_length ** 2) / moment_of_inertia)

# Damped natural frequency
def damped_natural_frequency(undamped_natural_frequency, damping_ratio):
    return undamped_natural_frequency * np.sqrt(1 / (1 - 2 * (damping_ratio ** 2)))

# Unforced response
def unforced_response(time, state, wn, zeta):
    theta, theta_dot = state
    dtheta_dt = theta_dot
    dtheta_dot_dt = -2 * zeta * wn * theta_dot - wn ** 2 * theta
    return [dtheta_dt, dtheta_dot_dt]

# Forced response
def forced_response(time, wm, wn, zeta):
    lambda_value = ((mass_system * pivot_distance * wm ** 2 * motor_arm_length) / moment_of_inertia) / (
        (wn ** 2) * np.sqrt((1 - (wm ** 2 / wn ** 2)) ** 2 + (2 * zeta * wm / wn) ** 2)
    )
    phi = np.arctan((2 * zeta * wm / wn) / (1 - (wm ** 2 / wn ** 2)))
    theta = lambda_value * np.sin(wm * time - phi)
    return theta

# Theoretical frequency response
def theoretical_frequency_response(wm_wn, wn, zeta):
    magnitude_theta_data = []
    for wm in wm_wn:
        phi = np.arctan(2 * zeta * wm / (1 - wm ** 2 + 1e-10))
        theta = ((mass_system * pivot_distance * (wm * wn) ** 2 * motor_arm_length) / moment_of_inertia) / (
            wn ** 2 * np.sqrt((1 - wm ** 2) ** 2 + (2 * zeta * wm) ** 2)
        )
        magnitude_theta = np.abs(theta)
        magnitude_theta_data.append(magnitude_theta)
    return magnitude_theta_data

# Time span for the simulation
time_span = (0, 1)
time_values = np.linspace(*time_span, 1000)

# Create a figure for the plots
plt.figure(figsize=(16, 12))

# Plot unforced response for different damping ratios
plt.subplot(3, 2, 1)
for zeta_value in [0, 0.1, 1, 1.5]:
    solution = solve_ivp(
        unforced_response, time_span, [1, 0], args=(undamped_natural_frequency, zeta_value), t_eval=time_values
    )
    plt.plot(solution.t, solution.y[0], label=f'Damping Ratio = {zeta_value}')

plt.xlabel('Time (t)')
plt.ylabel('θ(t)')
plt.title('Unforced Response to an Initial Displacement')
plt.legend()
plt.grid(True)

# Plot forced response for different rotational speeds (undamped)
plt.subplot(3, 2, 2)
for wm_value in [undamped_natural_frequency - 1, 50, 80]:
    theta_values = forced_response(time_values, wm_value, undamped_natural_frequency, 0)
    plt.plot(time_values, theta_values, label=f'Rotational Speed = {wm_value}')

plt.xlabel('Time (t)')
plt.ylabel('θ(t)')
plt.title('Forced Response (Undamped)')
plt.legend()
plt.grid(True)

# Plot forced response for different rotational speeds (underdamped)
plt.subplot(3, 2, 3)
for wm_value in [10, float(damped_natural_frequency(undamped_natural_frequency, 0.4)), 100]:
    theta_values = forced_response(time_values, wm_value, undamped_natural_frequency, 0.4)
    plt.plot(time_values, theta_values, label=f'Rotational Speed = {wm_value}')

plt.xlabel('Time (t)')
plt.ylabel('θ(t)')
plt.title('Forced Response (Underdamped)')
plt.legend()
plt.grid(True)

# Plot forced response for different rotational speeds (critically damped)
plt.subplot(3, 2, 4)
for wm_value in [10, 30, 50]:
    theta_values = forced_response(time_values, wm_value, undamped_natural_frequency, 1)
    plt.plot(time_values, theta_values, label=f'Rotational Speed = {wm_value}')

plt.xlabel('Time (t)')
plt.ylabel('θ(t)')
plt.title('Forced Response (Critically Damped)')
plt.legend()
plt.grid(True)

# Plot forced response for different rotational speeds (overdamped)
plt.subplot(3, 2, 5)
for wm_value in [10, 30, 50]:
    theta_values = forced_response(time_values, wm_value, undamped_natural_frequency, 1.5)
    plt.plot(time_values, theta_values, label=f'Rotational Speed = {wm_value}')

plt.xlabel('Time (t)')
plt.ylabel('θ(t)')
plt.title('Forced Response (Overdamped)')
plt.legend()
plt.grid(True)

# Plot theoretical frequency response for different damping ratios
plt.subplot(3, 2, 6)
wm_wn_values = np.linspace(0.1, 2.0, 100)
for zeta_value in [0.1, 0.2, 0.4, 0.8, 1, 1.2]:
    magnitude_theta_data = theoretical_frequency_response(wm_wn_values, undamped_natural_frequency, zeta_value)
    plt.plot(wm_wn_values, magnitude_theta_data, label=f'Damping Ratio={zeta_value}')

plt.xlabel('ωm/wn')
plt.ylabel('Magnitude of Theta')
plt.title('Theoretical Frequency Response for Different Damping Ratios')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
